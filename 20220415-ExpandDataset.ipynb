You're given a set of data that is aggregated on a monthly basis (as illustrated in Table A). Write code that can expand this monthly table into a daily table which spreads revenue across the 30 day period (as shown in Table B).  

You can assume each month has 30 days and that Table A is in a dataframe named `df`.

**Table A**  

|	| Month | Revenue | 
|:--|:--|:--| 
| 0	| 1	| 300 | 
| 1	| 2	| 330 |   
| 2	| 3	| 390 | 


**Table B**  

|	| Month | Day | Revenue | 
|:--|:--|:--| :--|
| 0	| 1	| 1 | 10 | 
| 1	| 1	| 2 | 10 |   
| 2	| 1	| 3 | 10 | 
| ... | ...	| ... | ... |
| 30 | 2 | 1 | 11 |
| 31 | 2 | 2 | 11 | 
| ... | ...	| ... | ... |
| 60 | 3 | 1 | 13 |
| ... | ...	| ... | ... |
| 89 | 3 | 30 | 13 |

$~$

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Month': [1, 2, 3],
    'Revenue': [300, 330, 390]
})
```

```python
df
```

```python
# list of days 1-30, repeated n times where n is the number of months in df
days = [i for i in range(1, 30+1)] * len(df['Month'])
```

```python
# list of months 1-n, where month i is repeated 30 times 
months = sum([[i] * 30 for i in df['Month']], [])
```

```python
# data frame linking months and days
months_days = pd.DataFrame({
    'Month': months,
    'Day': days
})
```

```python
months_days
```

```python
# merge data frame containing months and revenue with the data frame containing months and days
expanded_df = df.merge(months_days, on = 'Month')
```

```python
# convert revenue into the daily average by dividing by 30
expanded_df['Revenue'] = expanded_df['Revenue']/30
```

```python
# reorder the columns
expanded_df = expanded_df[['Month', 'Day', 'Revenue']]
```
